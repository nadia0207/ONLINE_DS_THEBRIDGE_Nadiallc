## Ejercicio 1
1. Elimina todos los VW de la siguiente lista. Crea una lista nueva para ello.
2. Hazlo de otra manera. Crea una lista nueva, y ve añadiendo todos los que no sean VW
3. En vez de `for`, usa un bucle `while`
> ejer_1 = ["VW", "Audi", "Renault", "VW", "BMW"]

```python
ejer_1 = ["VW", "Audi", "Renault", "VW","VW", "BMW"]
print("Lista inicial", ejer_1)
i = 0
nueva_lista = []

while i < len(ejer_1):
    if ejer_1[i] != "VW":
        nueva_lista.append(ejer_1[i])
    i += 1
print("Nueva lista ", nueva_lista)
        
```

```python
lista_2 = [] #Lista para guardar los numeros ingresado diferentes a 0
lista_pares = [] #Lista para guardar solo los numeros pares ingresado
suma = 0
i = 1

while i != 0:
    numero = int(input("Ingrese un numero")) 
    if numero != 0: lista_2.append(numero)
    else: break
    i = numero

for j in lista_2:
    if j % 2 == 0: lista_pares.append(j)

print(lista_2)
print(lista_pares)
print("El sumatorio de todos los enteros pares fue de",sum(lista_pares))
print("Los numeros pares fueron", lista_pares)

```

## Ejercicio 4
Escribir un programa que vaya pidiendo la cantidad de compras de un cliente. El programa tiene que pedir cantidades hasta que se ingresa un 0. Igualmente cuando una de las cantiaddes introducida es negativa, hay que mandar un mensaje de que "la cantidad es negativa, por favor introducir numeros positivos"

```python
lista_cantidades = [] #Lista para guardar los numeros ingresado diferentes a 0
suma_cantidades = 0
m = 1

while m != 0:
    cantidad_compra = int(input("Ingrese cantidad a comprar")) 
    if cantidad_compra < 0:
        print("La cantidad es negativa, por favor introducir numeros positivos")
        continue
    elif cantidad_compra != 0: lista_cantidades.append(cantidad_compra)
    else: break
    m = cantidad_compra

print(lista_cantidades)
print(f"El sumatorio de todas las cantidades es {sum(lista_cantidades)}")

```

## Ejercicio 6
Escribe un programa que calcule la cantidad de veces que está la letra "m" o "M" en la frase:
> "En un lugar de La Mancha, de cuyo nombre no quiero acordarme"

```python
frase = "En un lugar de La Mancha, de cuyo nombre no quiero acordarme"
cantidad_m = 0

for letra in frase:
    if letra == "m" or letra == "M": cantidad_m = cantidad_m + 1

print("Numero de veces que aperece \"m\" o \"M\" es", cantidad_m) 
```

## Ejercicio 11
Suponiendo que siempre se cumple que las dos listas del ejercicio son iguales de tamaño, implementa un programa que calcule el ratio ingreso/gasto de cada elemento, y lo guarde en una nueva lista. Si hubiese algún problema con los datos, hay que almacenar un mensaje de error en la lista del resultado. Usa `try/except`
> ingresos = [100, 200, 500, 100, 600]
>
> gastos = [50, 20, 70, 0, 25]

```python
try:
    ingresos = [100, 200, 500, 100, 600]
    gastos = [50, 20, 70, 0, 25]
    ratio = []
    
    for index, name in enumerate(ingresos):
        ratio. append((name / gastos[index]))
    print(ratio)

except Exception as e:
    print(e) #Se imprime el error
    print('Codigo con error')
```
