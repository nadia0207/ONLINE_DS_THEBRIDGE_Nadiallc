## Ejercicio 2
Eleva todos los elementos de la lista al cuadrado
> ```ejer_2 = [1,2,3,4,5]```

````python
ejer_2 = [1,2,3,4,5,8]
nueva_ejer_2 = [] #Lista para guardar los valores al cuadrado

for num_lista in ejer_2: #Recorrer la lista ejer_2 
    nueva_ejer_2.append(num_lista * num_lista) #Agregando un elemento a la nueva lista 

print(nueva_ejer_2)
````

## Ejercicio 3
Crea una lista nueva con todas las combinaciones de las siguientes dos listas:
> ```ejer_3_1 = ["Hola", "amigo"]```
>
> ```ejer_3_2 = ["Que", "tal"]```

Obten el siguiente output:

['Hola Que', 'Hola tal', 'amigo Que', 'amigo tal']

````python
ejer_3_1 = ["Hola", "amigo"]
ejer_3_2 = ["Que", "tal"]
ejer_3_nueva = [] 

for letra_1 in ejer_3_1: #Recorro la primera lista
    for letra_2 in ejer_3_2: #Recorro la segunda lista 
        ejer_3_nueva.append(letra_1 + " " + letra_2) #Agrego los valores obtenidos en el recorrido de cada lista 

print(ejer_3_nueva)
````

## Ejercicio 5
Dada la siguiente lista, elimina todos los valores iguales a 3
> ```ejer_5 = [3, 20, 3, 47, 19, 3, 29, 45, 67, 78, 90, 3, 3]```

Consejo: No intentes eliminar elementos sobre la lista que estás iterando. Haz una copia con `ejer_5.copy()`.

````python
ejer_5 = [3, 20, 3, 47, 19, 3, 29, 45, 67, 78, 90, 3, 3]
ejer_5_copy = ejer_5.copy() #Copio la lista ejer_5

for num_5 in ejer_5:  #Recorro valor por valor la lista ejer_5
    if num_5 == 3:   
        ejer_5_copy.remove(num_5)  #Elimino el valor de la lista copiada para que no afecte el recorrido de la lista original

print(ejer_5_copy)
````

## Ejercicio 9
1. ¿Cuántas veces se repite el 3 en la siguiente tupla?
2. Crea una tupla nueva con los elementos desde la posicion 5 a la 10.
3. ¿Cuántos elementos tiene la tupla `ejer_9`?
> ```ejer_9 = (3, 20, 3, 47, 19, 3, 29, 45, 67, 78, 90, 3, 3, 5, 2, 4, 7, 9, 4, 2, 4, 3, 3, 4, 6, 7)```

````python
ejer_9 = (3, 20, 3, 47, 19, 3, 29, 45, 67, 78, 90, 3, 3, 5, 2, 4, 7, 9, 4, 2, 4, 3, 3, 4, 6, 7)
contador = 0
nueva_tupla_ejer9 = ejer_9[4:10] #Creo una nueva tupla de la tupla anterior desde la posicion 5 hasta la posicion 10

for num_ejer_9 in ejer_9:  #Recorre valor por valor la tupla
    if  num_ejer_9 == 3: contador += 1 #Si el valor obtenido en el recorrido es igual a 3 sumamos 1 al contador

print("Numero de veces que se repite el numero 3 es:",contador)
print("Nueva tupla con valores desde posicion 5 a la 10:", nueva_tupla_ejer9)
print("Cantidad de elementos de la Tupla inicial:",len(ejer_9))
````

````python
ejer_9_1 = (3, 20, 3, 47, 19, 3, 29, 45, 67, 78, 90, 3, 3, 5, 2, 4, 7, 9, 4, 2, 4, 3, 3, 4, 6, 7)
veces_numero3 = ejer_9_1.count(3)
print("Numero de veces que se repite el numero 3 es:",veces_numero3)
````
