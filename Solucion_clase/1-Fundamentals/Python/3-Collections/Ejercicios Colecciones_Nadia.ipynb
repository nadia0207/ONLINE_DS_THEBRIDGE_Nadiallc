## Ejercicio 3
Crea una lista nueva con todas las combinaciones de las siguientes dos listas:
> ```ejer_3_1 = ["Hola", "amigo"]```
>
> ```ejer_3_2 = ["Que", "tal"]```

Obten el siguiente output:

['Hola Que', 'Hola tal', 'amigo Que', 'amigo tal']

````python
ejer_3_1 = ["Hola", "amigo"]
ejer_3_2 = ["Que", "tal"]
ejer_3_nueva = [] 

for letra_1 in ejer_3_1: #Recorro la primera lista
    for letra_2 in ejer_3_2: #Recorro la segunda lista 
        ejer_3_nueva.append(letra_1 + " " + letra_2) #Agrego los valores obtenidos en el recorrido de cada lista 

print(ejer_3_nueva)
````

````python
ejer_5 = [3, 20, 3, 47, 19, 3, 29, 45, 67, 78, 90, 3, 3]
ejer_5_copy = ejer_5.copy() #Copio la lista ejer_5

for num_5 in ejer_5:  #Recorro valor por valor la lista ejer_5
    if num_5 == 3:   
        ejer_5_copy.remove(num_5)  #Elimino el valor de la lista copiada para que no afecte el recorrido de la lista original

print(ejer_5_copy)
````

## Ejercicio 12
Convierte la siguiente tupla en un diccionario
> ```ejer_12 = [("x", 1), ("x", 2), ("x", 3), ("y", 1), ("y", 2), ("z", 1)]```

TIP: Puedes usar dos variables de control en un bucle `for`. Una por cada valor de las tuplas.

````python
ejer_12 = [("x", 1), ("x", 2), ("x", 3), ("y", 1), ("y", 2), ("z", 1),("x", 8)]
dicc_ejer_12 = {}
acumulador = 0
clave = ""
valor_12 = 0

for valor_lista in ejer_12: #("x", 1)
    for valor_tupla_12 in valor_lista: #"x", 1
        if acumulador == 0: 
            clave = valor_tupla_12
            acumulador += 1
            continue
        else:
            valor_12 = valor_tupla_12
        
        dicc_ejer_12[clave] = valor_12
        acumulador = 0
    
    clave = ""
    valor_12 = 0
    
print(dicc_ejer_12)
        

````

## Ejercicio 13

2. Crea otra lista ordenada descendente con los valores
3. Añade una nueva clave/valor
4. Busca la clave 2 dentro del diccionario
5. Itera la clave y el valor del diccionario con un unico for
> ```ejer_13 = {4:78, 2:98, 8:234, 5:29}```

````python
ejer_13 = {4:78, 2:98, 8:234, 5:29}

#Crea una lista ordenada ascendente con las claves del diccionario
lista_13_claves = list(ejer_13.keys())
lista_13_claves.sort() #no devuelve nada (None), modifica la lista original 
print(lista_13_claves)

#Crea otra lista ordenada descendente con los valores
lista_13_valores = list(ejer_13.values())
lista_13_valores.sort(reverse = True)
print(lista_13_valores)

#Añade una nueva clave/valor
ejer_13[10] = 35
print(ejer_13)

#Busca la clave 2 dentro del diccionario
print(ejer_13.get(2)) #Aqui muestra el valor del dato encontrado en la clave 2
print(2 in ejer_13)

#Itera la clave y el valor del diccionario con un unico for
for i in ejer_13:
    print(f"clave--> {i}, Valor --> {ejer_13[i]}")

````

## Ejercicio 16
Multiplica todos los valores del diccionario
> ```ejer_16 = {1: 11, 2: 22, 3: 33, 4: 44, 5: 55}```

````python
ejer_16 = {1: 11, 2: 22, 3: 33, 4: 44, 5: 55}
multi_16 = 1

for valor_16 in ejer_16:
    multi_16 *= ejer_16[valor_16] #multi_16 = multi_16 * ejer_16[valor_16]

print("La multiplicacion de todos los valores es:", multi_16)
````
