## Ejercicio 1
Escribe una función que convierta números del 1 al 7 en nombres de los dias de la semana. La función constará de un único argumento numérico y una salida de tipo string

````python
def dias_semana(numerodia):
    #dias_semana = ""
    match numerodia:
        case 1: dia_descripcion = "Lunes"
        case 2: dia_descripcion = "Martes"
        case 3: dia_descripcion = "Miercoles"
        case 4: dia_descripcion = "Jueves"
        case 5: dia_descripcion = "Viernes"
        case 6: dia_descripcion = "Sabado"
        case 7: dia_descripcion = "Domingo"
        case _: dia_descripcion = "Valor ingresado es incorrecto, día no existe"

    return dia_descripcion
````

````python
dias_semana(5)
````

````python
dias_semana(32)
````

## Ejercicio 2
En el ejercicio 8 de bucles, creábamos una pirámide invertida, cuyo número de pisos venía determinado por un input del usuario. Crear una función que replique el comportamiento de la pirámide, y utiliza un único parámetro de entrada de la función para determinar el número de filas de la pirámide, es decir, elimina la sentencia input.

````python

#cantidad_filas = int(input("Ingrese un numero de pisos"))
def imprime_pirarmide(pisos):
    for i in range(pisos,0,-1): 
        for j in range (1,i + 1): 
            print(j, end= " ")
        print()

````

## Ejercicio 3
Escibe una función que compare dos números. La función tiene dos argumentos y hay tres salidas posibles: que sean iguales, que el primero se  mayor que el segundo, o que el segundo sea mayor que el primero

````python
def compara_numeros(numero1, numero2):
    if(numero1 == numero2): resultado = "Numeros iguales"
    elif (numero1 > numero2): resultado = f"Primer numero ingresado {numero1}" +\
                                          f" es mayor que el segundo numero {numero2} "
    else: resultado = f"Primer numero ingresado {numero1}" +\
                                          f" es menor  que el segundo numero {numero2} "
    return print(resultado)

````

````python
compara_numeros(0, 3)
````

## Ejercicio 4
Escribe una función que sea un contador de letras. En el primer argumento tienes que introducir un texto, y el segundo que sea la letra a contar. La función tiene que devolver un entero con el número de veces que aparece esa letra, tanto mayuscula, como minúscula

````python

def contador_letras(texto, letra):
    contador = texto.lower().count(letra)
    
    return contador
````

````python
contador_letras("Bienvenido, hi","i")
````
