## Ejercicio 1
Escribe una función que convierta números del 1 al 7 en nombres de los dias de la semana. La función constará de un único argumento numérico y una salida de tipo string

````python
def dias_semana(numerodia):
    #dias_semana = ""
    match numerodia:
        case 1: dia_descripcion = "Lunes"
        case 2: dia_descripcion = "Martes"
        case 3: dia_descripcion = "Miercoles"
        case 4: dia_descripcion = "Jueves"
        case 5: dia_descripcion = "Viernes"
        case 6: dia_descripcion = "Sabado"
        case 7: dia_descripcion = "Domingo"
        case _: dia_descripcion = "Valor ingresado es incorrecto, día no existe"

    return dia_descripcion
````

````python
dias_semana(5)
````

````python
dias_semana(32)
````

## Ejercicio 3
Escibe una función que compare dos números. La función tiene dos argumentos y hay tres salidas posibles: que sean iguales, que el primero se  mayor que el segundo, o que el segundo sea mayor que el primero

````python
def compara_numeros(numero1, numero2):
    if(numero1 == numero2): resultado = "Numeros iguales"
    elif (numero1 > numero2): resultado = f"Primer numero ingresado {numero1}" +\
                                          f" es mayor que el segundo numero {numero2} "
    else: resultado = f"Primer numero ingresado {numero1}" +\
                                          f" es menor  que el segundo numero {numero2} "
    return print(resultado)

````

````python
compara_numeros(7, 3)
````

## Ejercicio 5
Escribe una función que tenga un único argumento, un string. La salida de la función tiene que ser un diccionario con el conteo de todas las letras de ese string.

````python
def convierte_a_diccionario(texto_5): #"Hola como Estas h"
    dicc_5 = {}

    for i in texto_5: #['H', 'o', 'l', 'a', ' ', 'c', 'o', 'm', 'o', ' ', 'E', 's', 't', 'a', 's',' ','h']
        i = i.lower()
        if i in dicc_5:
            dicc_5[i] += 1
        else:
            dicc_5[i] = 1

    return dicc_5
````

````python
print(convierte_a_diccionario("Hola como Estas h"))
print(convierte_a_diccionario("Buenos TARDES "))
````

## Ejercicio 7
Crea una función que reciba un número arbitrario de palabras, y devuelva una frase completa, separando las palabras con espacios.

````python
def formar_frase(*palabras):
    frase = ""
    for i in palabras:
        frase = frase +" "+ i

    return frase

````

## Ejercicio 8
Escribe un programa que obtenga el enésimo número de la [serie de Fibonacci](https://es.wikipedia.org/wiki/Sucesi%C3%B3n_de_Fibonacci). Tienes que crear una función recursiva con un único argumento.

```Python
fibonacci(2) = 1
fibonacci(4) = 3
fibonacci(5) = 5
fibonacci(8) = 21
```

````python
def fibonacci(numero): #2
    
    if numero == 0:
        acumulador = 0
        return acumulador
    elif numero == 1:
        acumulador = 1
        return acumulador
    else:    
        return (fibonacci(numero - 1) + fibonacci(numero - 2)) #fibonacci(1) + fibonacci(0) = 1 + 0
    
    #fibonacci(0) = 0
    #fibonacci(1) = 1
    #fibonacci(2) = fibonacci(1) + fibonacci(0) = 1 + 0 = 1
    #fibonacci(3) = fibonacci(2) + fibonacci(1) = 1 + 1 = 2
    #fibonacci(4) = fibonacci(3) + fibonacci(2) = 2 + 1 = 3
    #fibonacci(5) = fibonacci(4) + fibonacci(3) = 3 + 2 = 5
````

````python
print(fibonacci(2))
print(fibonacci(4))
print(fibonacci(5))
print(fibonacci(8))
````

## Ejercicio 9
Define en una única celda las siguientes funciones:
* Función que calcule el área de un cuadrado
* Función que calcule el area de un triángulo
* Función que calcule el área de un círculo

En otra celda, calcular el area de:
* Dos círculos de radio 10 + un triángulo de base 3 y altura 7
* Un cuadrado de lado = 10 + 3 círculos (uno de radio = 4 y los otros dos de radio = 6) + 5 triángulos de base = 2 + altura = 4

````python
import math

def calcular_area_cuadrado(lado):
    area_cuadrado = lado ** 2
    return area_cuadrado

def calcular_area_triangulo(base, altura):
    area_triangulo = (base * altura)/2
    return area_triangulo

def calcular_area_circulo(radio):
    area_circulo = (math.pi * radio ** 2)
    return area_circulo
````

````python
#Dos círculos de radio 10 + un triángulo de base 3 y altura 7
print(2*(calcular_area_circulo(10)) + calcular_area_triangulo(3,7))
````

````python
#Un cuadrado de lado = 10 + 3 círculos (uno de radio = 4 y los otros dos de radio = 6) + 5 triángulos de base = 2 + altura = 4
#Un cuadrado de lado = 10 
cuadrados = calcular_area_cuadrado(10)

#3 círculos (uno de radio = 4 y los otros dos de radio = 6)
circulos = calcular_area_circulo(4) + (2 *calcular_area_circulo(6))

#5 triángulos de base = 2 + altura = 4
triangulos = 5 * calcular_area_triangulo(2,4)

print(cuadrados + circulos + triangulos)
````
